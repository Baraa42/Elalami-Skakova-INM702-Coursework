### Part 3 -    Prepating Data for use and  functions



```python
from torch.utils.data import DataLoader
batch_size = 128
train_loader = DataLoader(dataset=train_dataset,
                          batch_size=batch_size,
                          shuffle=True)
test_loader = DataLoader(dataset=test_dataset,
                          batch_size=batch_size,
                          shuffle=False)
```

```python
def run_ANN(model, criterion, optimizer, train_loader, test_loader, epochs=10):
    '''
    Runs the model and returns the list of train/test losses/accuracies
    '''
    # Losses
    train_losses = np.zeros(epochs)
    test_losses = np.zeros(epochs)
    train_accuracy = np.zeros(epochs)
    test_accuracy = np.zeros(epochs)
    
    for i in range(epochs):
        train_loss =[]
        correct = 0 
        total = 0
        for inputs, targets in train_loader :

            inputs, targets = inputs.to(device), targets.to(device)
            # Reshaping Inputs to feed for model
            inputs = inputs.view(-1, D)
            # Resetting gradient
            optimizer.zero_grad()
            # Getting output and Loss
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            # Back prop
            loss.backward()
            optimizer.step()
            # Train loss after one minibatch
            train_loss.append(loss.item())
            # Getting prediction
            _, predictions = torch.max(outputs, 1)
            # Getting number correct prediction
            correct += (predictions==targets).sum().item()
            total += targets.shape[0]
        # Train loss after epoch and accuracy
        train_loss = np.mean(train_loss)
        train_losses[i] = train_loss
        train_accuracy[i]= correct/total
        
        test_loss =[]
        correct = 0 
        total = 0
        for inputs, targets in test_loader:

            inputs, targets = inputs.to(device), targets.to(device)
            # Reshaping Inputs to feed for model
            inputs = inputs.view(-1, D)
            #Getting output and Loss
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            # Test loss after one minibatch
            test_loss.append(loss.item())
            # Getting prediction
            _, predictions = torch.max(outputs, 1)
            # Getting number correct prediction
            correct += (predictions==targets).sum().item()
            total += targets.shape[0]
        # Test loss after epoch  and accuracy 
        test_loss = np.mean(test_loss)
        test_losses[i] = test_loss 
        test_accuracy[i]= correct/total
         
        print(f'Epochs {i+1}/{epochs}, Train Loss: {train_loss:.3f}, Test Loss : {test_loss:.3f}, Train Accuracy: {train_accuracy[i]:.3f}, Test Accuracy : {test_accuracy[i]:.3f}')
        
    return train_losses, test_losses, train_accuracy, test_accuracy
```

```python
def run_CNN(model, criterion, optimizer, train_loader, test_loader, epochs=10):
    '''
    Runs the model and returns the list of train/test losses/accuracies
    '''
    # Losses
    train_losses = np.zeros(epochs)
    test_losses = np.zeros(epochs)
    train_accuracy = np.zeros(epochs)
    test_accuracy = np.zeros(epochs)
    
    for i in range(epochs):
        train_loss =[]
        correct = 0 
        total = 0
        for inputs, targets in train_loader :

            inputs, targets = inputs.to(device), targets.to(device)
            # Resetting gradient
            optimizer.zero_grad()
            # Getting output and Loss
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            # Back prop
            loss.backward()
            optimizer.step()
            # Train loss after one minibatch
            train_loss.append(loss.item())
            # Getting prediction
            _, predictions = torch.max(outputs, 1)
            # Getting number correct prediction
            correct += (predictions==targets).sum().item()
            total += targets.shape[0]
        # Train loss after epoch and accuracy
        train_loss = np.mean(train_loss)
        train_losses[i] = train_loss
        train_accuracy[i]= correct/total
        
        test_loss =[]
        correct = 0 
        total = 0
        for inputs, targets in test_loader:

            inputs, targets = inputs.to(device), targets.to(device)
            #Getting output and Loss
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            # Test loss after one minibatch
            test_loss.append(loss.item())
            # Getting prediction
            _, predictions = torch.max(outputs, 1)
            # Getting number correct prediction
            correct += (predictions==targets).sum().item()
            total += targets.shape[0]
        # Test loss after epoch  and accuracy 
        test_loss = np.mean(test_loss)
        test_losses[i] = test_loss 
        test_accuracy[i]= correct/total
         
        print(f'Epochs {i+1}/{epochs}, Train Loss: {train_loss:.3f}, Test Loss : {test_loss:.3f}, Train Accuracy: {train_accuracy[i]:.3f}, Test Accuracy : {test_accuracy[i]:.3f}')
        
    return train_losses, test_losses, train_accuracy, test_accuracy
```

```python
train_losses, test_losses, train_accuracy, test_accuracy = run_ANN(model, criterion, optimizer, train_loader, test_loader, epochs=30)
```

### Part 5 -   Building CNN with pytorch



```python
K = 10
class CNN(nn.Module):
    def __init__(self, K):
        super(CNN, self).__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_channels=3, out_channels=32, kernel_size=3, stride=2),
            nn.ReLU(),
            nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, stride=2),
            nn.ReLU(),
            nn.Conv2d(in_channels=64, out_channels=128, kernel_size=3, stride=2),
            nn.ReLU(),
        )
        
        self.dense = nn.Sequential(
            nn.Dropout(0.2),
            nn.Linear(128*3*3, 512),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(512, K))
        
    def forward(self, X):
        out = self.conv(X)
        out = out.view(out.size(0),-1)
        out = self.dense(out)
        return out
```

```python
model = CNN(K)
print(model)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters())
```

### Part 6 -   New CNN



```python
K = 10
class CNN(nn.Module):
    def __init__(self, K):
        super(CNN, self).__init__()
        
        # conv layers
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, stride=2)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, stride=2)
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, stride=2)
        
        # liner layers
        self.fc1 = nn.Linear(128*3*3, 256)
        self.fc2 = nn.Linear(256, K)
        
        
    def forward(self, X):
        out = F.relu(self.conv1(X))
        out = F.relu(self.conv2(out))
        out = F.relu(self.conv3(out))
        out = out.view(-1, 128*3*3)
        out = F.dropout(out, p=0.2)
        out = F.relu(self.fc1(out))
        out = F.dropout(out, p=0.2)
        out = F.relu(self.fc2(out))
        return out
```

```python
model = CNN(K)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters())
```

## Training

```python
train_losses, test_losses, train_accuracy, test_accuracy = run_CNN(model, criterion, optimizer, train_loader, test_loader, epochs=15)
```

### Part 7 -   Improving CNN results using Google Colab GPU



In this section we make use of GPU to try obtaining better accuracy by increasing the number of convolutional layers. Since we couldn't manage to access Hyperion we used Google Colabs GPU and uploaded the notebook to the github repo.

The CNN architecture used is the following :

```python
K = 10
class CNN(nn.Module):
    def __init__(self, K):
        super(CNN, self).__init__()
        
        # conv layers
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=(1,1))
        self.conv2 = nn.Conv2d(32, 32, kernel_size=3, padding=(1,1))
    
        self.conv3 = nn.Conv2d(32, 64, kernel_size=3, padding=(1,1))
        self.conv4 = nn.Conv2d(64, 64, kernel_size=3, padding=(1,1))
        
        self.conv5 = nn.Conv2d(64, 128, kernel_size=3, padding=(1,1))
        self.conv6 = nn.Conv2d(128, 128, kernel_size=3, padding=(1,1))
        
        # liner layers
        self.fc1 = nn.Linear(128*4*4, 1024)
        self.fc2 = nn.Linear(1024, K)
        
        
    def forward(self, X):
        m = nn.MaxPool2d(2)
        out = F.relu(self.conv1(X))
        out = F.relu(self.conv2(out))
        out = m(out)
        out = F.relu(self.conv3(out))
        out = F.relu(self.conv4(out))
        out = m(out)
        out = F.relu(self.conv5(out))
        out = F.relu(self.conv6(out))
        out = m(out)
        out = out.view(-1, 128*4*4)
        out = F.dropout(out, p=0.2)
        out = F.relu(self.fc1(out))
        out = F.dropout(out, p=0.2)
        out = F.relu(self.fc2(out))
        return out
```

```python
model = CNN(K).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01, momentum=0.9)
```

```python
train_losses, test_losses, train_accuracy, test_accuracy = run_CNN(model, criterion, optimizer, train_loader, test_loader, epochs=15)
```
